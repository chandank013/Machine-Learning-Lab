### Naive Baye's Implementation on tips dataset

```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
print(tips.head())
```

```python
tips["tip_high"] = (tips["tip"] >= 3).astype(int)   # target column
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

tips["sex"] = encoder.fit_transform(tips["sex"])
tips["smoker"] = encoder.fit_transform(tips["smoker"])
tips["day"] = encoder.fit_transform(tips["day"])
tips["time"] = encoder.fit_transform(tips["time"])

print(tips.head())
```

```python
X = tips[["total_bill", "size", "sex", "smoker", "day", "time"]]  # features
y = tips["tip_high"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
X_train
```

```python

from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
y_pred = nb.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

#### Here smoker is target features

```python
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix

# Load dataset
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
# Features (excluding smoker) and Target (smoker)
X = tips.drop("smoker", axis=1)
y = tips["smoker"]
```

```python
# Encode categorical features
X = pd.get_dummies(X, drop_first=True)
X
```

```python
# Encode target (smoker yes/no)
le = LabelEncoder()
y = le.fit_transform(y)   # "No"=0, "Yes"=1
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
y_pred = nb.predict(X_test)

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=["No", "Yes"]))
```

```python
print("Accuracy:", nb.score(X_test, y_test))
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Heatmap Visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=nb.classes_, yticklabels=nb.classes_)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix Heatmap - Naive Bayes on Tips Dataset")
plt.show()

print(classification_report(y_test, y_pred))
```
