### 01. Importing required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 02. Load the wine dataset

```python
from sklearn.datasets import load_wine

wine_data = load_wine()
wine_data
```

```python
X = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
y = wine_data.target
```

### 03. Data Visualisation

```python
plt.figure(figsize=(12,8))
sns.heatmap(X.corr(), annot=False, cmap='coolwarm', center=0, 
            cbar_kws={'shrink': 0.8})
plt.title('Features Correlation Matrix', fontsize=12, fontweight='bold')
plt.show()
```

### 04. Data processing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
# Feature scaling 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
# Model Training
from sklearn.svm import SVC

svm_basic = SVC()
svm_basic.fit(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix

# Make predictions
y_pred = svm_basic.predict(X_test_scaled)
y_pred
```

```python
acc_score = accuracy_score(y_test, y_pred)
print("Accuracy_score of the model is: \n", acc_score)
con_mat1 = confusion_matrix(y_test, y_pred)
print("confusion_matrix of the model is: \n", con_mat1)
cls_rep = classification_report(y_test, y_pred)
print("Classification_report of the model is: \n", cls_rep)
```

### 05. Cross validation

```python
from sklearn.model_selection import GridSearchCV, cross_val_score

cv_scores = cross_val_score(svm_basic, X_train_scaled, y_train, cv=5, scoring='accuracy')
print("5-fold CV accuracy: ",cv_scores.mean() ,"Â±", cv_scores.std())
```

### 06. Hyperparameter Tuning

```python
param_grid_poly = [
    # RBF and sigmoid kernels
    {
        'C': [0.01, 0.1, 1, 10, 100, 1000],
        'kernel': ['rbf', 'sigmoid'],
        'gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1, 10]
    },
    # Polynomial kernel with degree parameter
    {
        'C': [0.01, 0.1, 1, 10, 100, 1000],
        'kernel': ['poly'],
        'gamma': ['scale', 'auto', 0.001, 0.01, 0.1, 1],
        'degree': [2, 3, 4]
    }
]
```

```python
# Perform grid search
grid_search = GridSearchCV(
    SVC(random_state=42),
    param_grid_poly,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=0  
)
```

```python
grid_search.fit(X_train_scaled, y_train)
```

```python
grid_search.cv_results_['params']
```

```python
print(f"\nBest parameters found:")
for param, value in grid_search.best_params_.items():
    print(f"  {param}: {value}")
```

```python
print("Best cross-validation score: ", grid_search.best_score_)
```

### 07. Final evaluation 

```python
# Get the best model
best_svm = grid_search.best_estimator_
best_svm
```

```python
# Make predictions on test set
y_pred_best = best_svm.predict(X_test_scaled)
print(y_pred_best)
```

```python
acc_score = accuracy_score(y_test, y_pred_best)
print("Accuracy_score of the model is: \n", acc_score)
con_mat2 = confusion_matrix(y_test, y_pred_best)
print("confusion_matrix of the model is: \n", con_mat2)
cls_rep = classification_report(y_test, y_pred_best)
print("Classification_report of the model is: \n", cls_rep)
```

```python
best_accuracy = accuracy_score(y_test, y_pred_best)
print("Best_accuracy in all models is: \n",best_accuracy)
```

```python
print(f"Improvement over baseline: {(best_accuracy - acc_score)*100:.2f} percentage points")
```

### 09. ONE-VS-ALL Analysis

```python
# Decision function scores (distances from hyperplane)
decision_scores = best_svm.decision_function(X_test_scaled)
print(f"\nDecision function output shape: {decision_scores.shape}")
print("Each sample gets 3 scores (one for each OvA classifier)")
```

```python
# Analyze decision confidence
max_scores = np.max(decision_scores, axis=1)
min_scores = np.min(decision_scores, axis=1)
confidence_margin = max_scores - min_scores

print(f"\nDecision confidence analysis:")
print(f"Average confidence margin: {confidence_margin.mean():.3f}")
print(f"Minimum confidence margin: {confidence_margin.min():.3f}")
print(f"Maximum confidence margin: {confidence_margin.max():.3f}")
```

```python
# Identify most and least confident predictions
most_confident_idx = np.argmax(confidence_margin)
least_confident_idx = np.argmin(confidence_margin)

print(f"\nMost confident prediction:")
print(f"  Sample index: {most_confident_idx}")
print(f"  Predicted class: {wine_data.target_names[y_pred_best[most_confident_idx]]}")
print(f"  Actual class: {wine_data.target_names[y_test[most_confident_idx]]}")
print(f"  Confidence margin: {confidence_margin[most_confident_idx]:.3f}")

print(f"\nLeast confident prediction:")
print(f"  Sample index: {least_confident_idx}")
print(f"  Predicted class: {wine_data.target_names[y_pred_best[least_confident_idx]]}")
print(f"  Actual class: {wine_data.target_names[y_test[least_confident_idx]]}")
print(f"  Confidence margin: {confidence_margin[least_confident_idx]:.3f}")
```
